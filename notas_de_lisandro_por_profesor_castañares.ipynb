```python
class Docente:
    def __init__(self, nombre):
        self.nombre = nombre

    def asignar_tarea(self, alumno, tarea):
        alumno.tareas_pendientes.append(tarea)
        print(f"{self.nombre} ha asignado la tarea '{tarea}' a {alumno.nombre}.")

    def corregir_tarea(self, alumno, tarea):
        if tarea in alumno.tareas_pendientes:
            alumno.tareas_pendientes.remove(tarea)
            alumno.tareas_corregidas.append(tarea)
            print(f"{self.nombre} ha corregido la tarea '{tarea}' de {alumno.nombre}.")
        else:
            print(f"{self.nombre} no puede corregir la tarea '{tarea}' de {alumno.nombre}, no fue asignada previamente.")

    def brindar_retroalimentacion(self, alumno):
        if alumno.tareas_corregidas:
            print(f"{self.nombre} está brindando retroalimentación a {alumno.nombre} sobre las tareas corregidas.")
            for tarea in alumno.tareas_corregidas:
                print(f"Tarea corregida: {tarea}")
        else:
            print(f"No hay tareas corregidas para brindar retroalimentación a {alumno.nombre}.")


class Alumno:
    def __init__(self, nombre):
        self.nombre = nombre
        self.tareas_pendientes = []
        self.tareas_corregidas = []

    def resolver_tarea(self, tarea):
        if tarea in self.tareas_pendientes:
            print(f"{self.nombre} está resolviendo la tarea '{tarea}'.")
        else:
            print(f"{self.nombre} no puede resolver la tarea '{tarea}', no fue asignada previamente.")

    def ver_tareas_corregidas(self):
        if self.tareas_corregidas:
            print(f"Tareas corregidas de {self.nombre}:")
            for tarea in self.tareas_corregidas:
                print(tarea)
        else:
            print(f"No hay tareas corregidas para mostrar de {self.nombre}.")


# Ejemplo de uso
docente1 = Docente("Profesor García")
alumno1 = Alumno("Juan")
alumno2 = Alumno("María")

docente1.asignar_tarea(alumno1, "Ejercicio de matemáticas")
docente1.asignar_tarea(alumno2, "Redacción de un ensayo")

alumno1.resolver_tarea("Ejercicio de matemáticas")
alumno2.resolver_tarea("Ejercicio de matemáticas")  # No puede resolver esta tarea, no fue asignada

docente1.corregir_tarea(alumno1, "Ejercicio de matemáticas")
docente1.corregir_tarea(alumno2, "Redacción de un ensayo")  # No puede corregir esta tarea, no fue asignada

alumno1.ver_tareas_corregidas()
docente1.brindar_retroalimentacion(alumno1)
docente1.brindar_retroalimentacion(alumno2)  # No hay tareas corregidas para este alumno


def ingresar_notas():
    alumnos = {}
    continuar = True

    while continuar:
        nombre = input("Ingrese el nombre del alumno: ")
        notas = []

        while True:
            nota = input("Ingrese una nota del alumno (o 'q' para terminar): ")

            if nota == 'q':
                break

            notas.append(float(nota))

        profesor = input("Ingrese el nombre de la Materia: ")
        alumnos[nombre] = {'notas': notas, 'profesor': profesor}

        opcion = input("¿Desea ingresar las notas de otro alumno? (s/n): ")

        if opcion.lower() == 'n':
            continuar = False

    return alumnos

def calcular_promedio(alumnos):
    promedios = {}

    for alumno, datos in alumnos.items():
        notas = datos['notas']
        promedio = sum(notas) / len(notas)
        promedios[alumno] = promedio

    return promedios

def determinar_estado(promedios):
    estados = {}

    for alumno, promedio in promedios.items():
        if promedio >= 5:
            estados[alumno] = 'Aprobado'
        else:
            estados[alumno] = 'Reprobado'

    return estados

def mostrar_promedios_y_estados(promedios, estados):
    print("Promedios de notas y estado por alumno:")

    for alumno, promedio in promedios.items():
        estado = estados[alumno]
        print(f"{alumno}: Promedio: {promedio}, Estado: {estado}")



 # Programa principal
alumnos = ingresar_notas()
promedios = calcular_promedio(alumnos)
estados = determinar_estado(promedios)
mostrar_promedios_y_estados(promedios, estados)


```
